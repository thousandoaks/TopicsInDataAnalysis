### This procedure is useful when dealing with really large twitter queries (in terms of datapoints).
#### the main disadvantage is that you don't get all the information (e.g. metadata). You can always further query twitter for those conversations of interest.

```python
# Setting Max number of Tweets
maxTweets = 1000
```

```python
minRetweets=50
```

```python
end_time=datetime.datetime.now()-datetime.timedelta(minutes=20)
start_time=end_time-datetime.timedelta(days = 2)
```

```python
query='(vacunas OR vacuna) lang:es min_retweets:'+str(minRetweets)+' since:'+start_time.strftime("%Y-%m-%d")+' until:'+end_time.strftime("%Y-%m-%d")
```

```python
query
```

```python
# Using TwitterSearchScraper to scrape data and append tweets to list
start_time = time.time()

DailyTweets = []


for i,tweet in enumerate(sntwitter.TwitterSearchScraper(query).get_items()):
    if i>maxTweets:
        break
    DailyTweets.append([tweet.date, tweet.id, tweet.content, tweet.user.username,tweet.retweetCount])

   
    
# Creating a dataframe from the tweets list above
DailyTweetsDataFrame = pd.DataFrame(DailyTweets, columns=['created_at', 'TweetId', 'text', 'author.name','public_metrics.retweet_count'])



print("--- %s seconds ---" % (time.time() - start_time))
```
