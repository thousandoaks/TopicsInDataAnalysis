### 2. Getting a user's Tweet timeline
#### In the previous lab we observe that @manuelmunizv was a top influencer. Let's investigate a bit further who he is.

#### Alternative 1: Fetch all the tweets he has posted in the last 7 days

```python
# This is where we specify our query 
query = "from:manuelmunizv "
```

```python
search_results = client.search_recent(query=query, max_results=100)
```

```python
ListofTweets=[]


limit = 50
counter=0
for page in search_results:
    
        
            
        # The Twitter API v2 returns the Tweet information and the user, media etc.  separately
        # so we use expansions.flatten to get all the information in a single JSON
        
        result = expansions.flatten(page)
        for tweet in result:    
            ListofTweets.append(tweet)
            
        
        
        
        # we set a counter to limit the number of API calls
        counter+=1
        if counter == limit:
               break
            
            
            
```

```python
ListofTweetsDataFrame=pd.json_normalize(ListofTweets)
```
